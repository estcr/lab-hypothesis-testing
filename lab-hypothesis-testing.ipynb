In this challenge, we will be working with pokemon data. The data can be found here:

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
from scipy.stats import ttest_ind


# Hipótesis:
# H0: HP_dragón ≤ HP_hierba
# H1: HP_dragón > HP_hierba

dragon_hp = df[df['Type 1'] == 'Dragon']['HP']
grass_hp = df[df['Type 1'] == 'Grass']['HP']

# Perform a one-tailed t-test
t_stat, p_value = ttest_ind(dragon_hp, grass_hp, alternative='greater')

# Significance level
alpha = 0.05

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Se rechaza la hipotesis nula. Los Pokemón de tipo Dragón poseen mas Hp que los de tipo Hierba.")
else:
    print("Fallo al rechazar hipotesis nula, NO hay una diferencia significante entre los HP de los Pokemón tipo Dragón e hierba.")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
from scipy.stats import ttest_ind

legendary_stats = df[df['Legendary'] == True]
non_legendary_stats = df[df['Legendary'] == False]

# Hipótesis:
# H0: Estadísticas entre Pokémon legendarios = Estadísticas entre Pokémon no legendarios.
# H1: Estadísticas entre Pokémon legendarios != Estadísticas entre Pokémon no legendarios.


# Prueba t para cada estadística
stats_to_test = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
results = {}

for stat in stats_to_test:
    t_stat, p_value = ttest_ind(legendary_stats[stat], non_legendary_stats[stat], equal_var=False)  # Asumimos varianzas distintas
    results[stat] = {'T-statistic': t_stat, 'P-value': p_value}

# Mostrar resultados
for stat, result in results.items():
    print(f"Stat: {stat}")
    print(f"  T-statistic: {result['T-statistic']:.4f}")
    print(f"  P-value: {result['P-value']:.4f}")
    if result['P-value'] < 0.05:
        print(f"  => Rechazamos H0: Diferencia significativa en {stat} entre Legendarios y No Legendarios.\n")
    else:
        print(f"  => No se rechaza H0: No hay diferencia significativa en {stat} entre Legendarios y No Legendarios.\n")
```

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df1.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import numpy as np
from scipy.stats import ttest_ind

# Coordenadas del colegio y el hospital
school_coords = (-118, 37)
hospital_coords = (-122, 34)

# Función para calcular la distancia euclidiana
def euclidean_distance(longitude, latitude, target_coords):
    return np.sqrt((longitude - target_coords[0])**2 + (latitude - target_coords[1])**2)

# Calcular distancias
df1['distance_to_school'] = df1.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], school_coords), axis=1)
df1['distance_to_hospital'] = df1.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], hospital_coords), axis=1)

# Clasificar las casas en cercanas y lejanas
df1['is_close'] = (df1['distance_to_school'] < 0.50) | (df1['distance_to_hospital'] < 0.50)

# División del dataset
close_houses = df1[df1['is_close']]
far_houses = df1[~df1['is_close']]

# Prueba T de hipótesis (asumimos varianzas distintas)
# H0: mu_cerca >= mu_lejos
# H1: mu_cerca < mu_lejos
t_stat, p_value = ttest_ind(close_houses['median_house_value'], far_houses['median_house_value'], equal_var=False, alternative = "less")

# Resultados
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretación
if p_value < 0.05:
    print("Rechazamos la hipótesis nula: Las casas cercanas son significativamente más caras.")
else:
    print("No podemos rechazar la hipótesis nula: No hay evidencia de que las casas cercanas sean más caras.")
```
